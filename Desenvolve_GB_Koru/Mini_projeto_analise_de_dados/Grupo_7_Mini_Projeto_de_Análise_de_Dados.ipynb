####Estruturas Condicionais e LaÃ§os

```python
 #Usar estrutura condicional (if/elif/else) com uma coluna numÃ©rica

valor= df['VALOR_VENDA'][0] #pega o valor da primeira venda da coluna VALOR_VENDA do df,[0] pega o primeiro elemento da coluna
if valor > 50000 :
  print("Suas vendas foram muito altas")
elif valor > 10000:
  print("As vendas ainda estÃ£o boas")
else:
  print("NÃ£o tÃ¡ tÃ£o boa mas ainda ta ok")
```

```python
#Usar um laÃ§o for para calcular a soma dos primeiros 5 valores da coluna 'VALOR_VENDA'
soma = 0
for valor in df['VALOR_VENDA'][:5]:
  soma += valor

print("A soma das 5 primeiras vendas Ã©:", soma)
```

```python
#Empregar um laÃ§o while para encontrar o primeiro valor da coluna 'VALOR_VENDA' que atenda a uma condiÃ§Ã£o
i = 0
while i < len(df['VALOR_VENDA']):
  if df['VALOR_VENDA'][i] > 9000:
    print("Primeira venda acima de R$ 9.000,00:",df['VALOR_VENDA'][i])
    break
  i += 1
```

#### Operadores AritmÃ©ticos e ManipulaÃ§Ã£o de Dados

 Soma

```python
#AdiÃ§Ã£o: Adicionar uma "taxa de processamento" fixa ao 'VALOR_VENDA'
TAXA_PROCESSAMENTO = 50.00 # Exemplo de valor fixo
df['VALOR_COM_TAXA'] = df['VALOR_VENDA'] + TAXA_PROCESSAMENTO
df.head()
```

<font color=cian> *Cria uma nova coluna com o valor depois de aplicaÃ§Ã£o de uma taxa de 50.00, assim mostrando o valor total com taxas aplicada*

 SubtraÃ§Ã£o

```python
#Cria uma nova coluna com o valor do desconto, que seria o preÃ§o sugerido pelo fabricante, menos o preÃ§o de revenda (PRECO_SUGERIDO - PRECO_UNITARIO).
df['DESCONTO'] = df['PRECO_SUGERIDO'] - df['PRECO_UNITARIO']

#Cria uma coluna para verificar se teve ou nÃ£o desconto na compra do produto, se o valor for maior que 0,
#Significa que o desconto foi aplicado, e a colunas mostra se Ã© True ou False, se recebeu ou nÃ£o desconto.
#TRUE Recebeu desconto
#FALSE NÃ£o recebeu desconto

df['COM_DESCONTO'] = df['DESCONTO'] > 0

df.head()
```

<font color=cian> *Cria a coluna desconto, utilizando o valor de preÃ§o sugerido pelo fabricante menos o valor do preÃ§o unitario que Ã© repassado para o consumidor, assim conseguimos saber se o produto foi vendido por um valor abaixo ou acima da mÃ©dia.*

<font color=cian> *Ãpos a criaÃ§Ã£o da coluna desconto foi criada outra coluna com desconto aplicado, e com o filtro que retorna um valor booleano, para verificar se o desconto foi aplicado ou nÃ£o.
TRUE se foi aplicado, e FALSE se nÃ£o foi aplicado.*

```python
#Cria a coluna COM_DESCONTO na base de dados.
df['COM_DESCONTO'].value_counts()
```

<font color=cian> *Visualiza quantos descontos foram aplicados, e quantos nÃ£o foram aplicados descontos.*

 MultiplicaÃ§Ã£o

```python
# MultiplicaÃ§Ã£o entre 'QUANTIDADE' e 'PRECO_UNITARIO'
# Esta operaÃ§Ã£o deve ser prÃ³xima ao valor de 'VALOR_VENDAS'
df['VENDAS_CALCULADAS'] = df['QUANTIDADE'] * df['PRECO_UNITARIO']
df.head()
```

####VisualizaÃ§Ã£o dos dados

```python
df['DATA_PEDIDO'] = pd.to_datetime(df['DATA_PEDIDO'])
df.groupby('DATA_PEDIDO')['VALOR_VENDA'].sum().plot(
    xlabel='Data Pedido',
    ylabel='NÃºmero de Vendas',
    title='NÃºmero de Vendas ao Longo do Tempo'
    )
```

<font color=cian> *AtravÃ©s do grÃ¡fico de linhas para anÃ¡lises temporais, Ã© possÃ­vel observar um padrÃ£o no final de ano (do mÃªs 10 ao mÃªs 12), perÃ­odo em que ocorre o maior volume de vendas do negÃ³cio.*

```python
df['LINHA_PRODUTO'].value_counts().plot(kind='bar')
plt.title('Produtos que mais venderam (LINHA_PRODUTO)')
plt.xlabel('Produtos')
plt.ylabel('NÃºmero de Pedidos')
plt.xticks(rotation=50)
plt.show()
```

<font color=cian> *Para visualizar dados categÃ³ricos, utilizou-se o grÃ¡fico de barras, que permite comparar a frequÃªncia e a quantidade de vendas de cada item, facilitando a identificaÃ§Ã£o dos produtos mais vendidos. Como observado, as linhas Classic Cars e Vintage Cars sÃ£o os produtos que apresentam o maior nÃºmero de vendas.*

<font color=cian> *Este grÃ¡fico apresenta a relaÃ§Ã£o entre as variÃ¡veis de QUANTIDADE e de VALOR_VENDA. Com ele, visualiza-se que quanto maior a quantidade vendida, maior Ã© o valor da venda (relaÃ§Ã£o positiva).*
